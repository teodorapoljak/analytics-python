We will just run the code to drop duplicates.

```python
df = df.drop_duplicates()
df
```

But if we want to actually save this dataframe to the variable we need to call the df variable after. 

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

First column that have issues is the Last_Name. We'll use strip command to remove everything extra from both sides of the text. As we see below, nothing has changed, below output that we got, default strip removes only spaces because we're not specifying any specific value.

```python
df['Last_Name'].str.strip()
```

Now we're going to specify the exact values and run it again.

```python
df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
df
```

Considering the various formats of phone numbers, we're firstly going to rid of all types of data format that is not a number.

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

Formating numbers as xxx-xxx-xxxx.

```python
df['Phone_Number'] = df['Phone_Number'].apply(
    lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] 
    if pd.notna(x) and len(x) == 10 
    else x
)
df
```

Replacing unknown values.
The 'NaN' value within Phone_Number is a float value, not the string and that is the reason why .str.replace('NaN', '') doesn't work (it only operates on string values). So we are going to try with another option.

```python
df['Phone_Number']=df['Phone_Number'].fillna('')
df
```

Replacing n/a cells as blank address.

```python
df['Address'].str.replace('N/a','')
```

Spliting combined information of street, state and zip code within Address column.

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand=True)
df
```

Because only the once that want to be contacted and have their phone number are of use in this database, we want to remove the others.
First we want to disregard the once that stated that they don't want to be contacted.

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace=True)
df
```

After we filtered out only the customers that want to be contacted, we are also going to filter only the customers whose phone numbers we have.

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Na','')
df
```

Droping that one row, that doesn't have either last name, phone number or want to be contacted.

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
df
```
